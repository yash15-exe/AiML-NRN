Using Logistic Regression model(Simple Machine Learning)

```python
import nbformat


with open('hand_digit_recognition.ipynb', 'r', encoding='utf-8') as f:
    notebook = nbformat.read(f, as_version=4)

with open('hand_digit_recognition.ipynb', 'w', encoding='utf-8') as f:
    nbformat.write(notebook, f)
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
mnist = fetch_openml("mnist_784")
```

```python
X, y = mnist.data/255, mnist.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train
```

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train, y_train)
```

```python
predicted_result = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predicted_result)
```

```python
accuracy
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_test, predicted_result)
```

```python
display = ConfusionMatrixDisplay(confusion_matrix = cm)
```

```python
display.plot(cmap='Reds')
plt.title('Confusion Matrix - Traditional ML')
plt.show()
```

Hand Digit recognition(Deep Neural Networks -TensorFlow)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
model = Sequential([Flatten(),
                Dense(64, activation = 'relu'),
                    Dense(10, activation = 'softmax')])
                   
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
X_train_scalled = X_train/255
```

```python
model.fit(X_train_scalled, y_train, epochs = 10)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
```

```python
import numpy as np
predicted_probs = model.predict(X_test/255)
predicted_output = np.argmax(predicted_probs, axis=1)

# Generate and display the confusion matrix
dp_com = confusion_matrix(y_test, predicted_output)
disp = ConfusionMatrixDisplay(confusion_matrix=dp_com)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix - Deep Learning Model")
plt.show()
```
